```r
pacman::p_load(tidyverse, stringr, stringi, ds4psy, janitor)
```

```r
USE_SAMPLE <- FALSE
```

```r
clean <- function(data) {
    str_replace_all(data, pattern = "\\[|\\]| ", replacement = "")
}
```

```r
encoding <- guess_encoding("data/day5_crates_test", n_max = 1000) |>
    filter(confidence == max(confidence)) |>
    pull(encoding)
```

```r
if (USE_SAMPLE) {
    ext <- "_test"
} else {
    ext <- ""
}

crates <-
    readLines(paste0("data/day5_crates", ext), skipNul = TRUE, encoding = encoding) |>
    gsub(pattern = "\\\xfe|\\\xff|[0-9]", replacement = "", ~.) |>
    gsub(pattern = "    ", replacement = " ", ~.)
crates <-
    read.table(text = crates, sep = " ", fill = TRUE) |>
    mutate_all(na_if, "") |>
    janitor::remove_empty("rows") |>
    rename_with(~ gsub("V", "", .x))
moves <-
    read.delim(paste0("data/day5_moves", ext), header = FALSE, sep = " ") |>
    select_if(is.integer) |>
    rename_with(~ c("move", "from", "to"))

crates
moves
```

```r
convert_to_vectors <- function(df) {
    vector <- as.array(df)
    vector[!is.na(vector)]
}
crates_array_in <- lapply(crates, convert_to_vectors)
```

```r
move_box <- function(crates_array, how_many_to_move, from, to) {
    crates_to_move <- crates_array[[from]][1:how_many_to_move]
    print(paste("crates_to_move", crates_to_move))
    print("from_before")
    print(crates_array[[from]])
    print(paste("removed crates", crates_array[[from]][1:how_many_to_move]))
    crates_array[[from]] <- crates_array[[from]][-1:-how_many_to_move]
    print("from_after")
    print(crates_array[[from]])
    print("to_before")
    print(crates_array[[to]])
    print("to_after")
    print(append(crates_array[[to]], crates_to_move, after = 0))
    crates_array[[to]] <- append(crates_array[[to]], crates_to_move, after = 0)
    return(crates_array)
}
```

```r
crates_array <- crates_array_in
move_boxes <- function(moves, crates_array, style = "sequential") {
    if (style == "sequential") {
        for (i in 1:nrow(moves)) {
            print("move")
            print(i)
            how_many <- moves[[i, 1]]
            print(paste("how_many",how_many))
            from <- moves[[i, 2]]
            to <- moves[[i, 3]]
            for (iterations in seq(1,how_many)) {
                print(paste("iterations",iterations))
                crates_array <- move_box(crates_array, 1, from, to)
            }
        }
    } else {
        print("all in one")
        for (i in 1:nrow(moves)) {
            print(paste("i",i))
            how_many <- moves[[i, 1]]
            from <- moves[[i, 2]]
            to <- moves[[i, 3]]
            print(paste("how_many",how_many))
            crates_array <- move_box(crates_array, how_many, from, to)
            
        }
    }
    return(crates_array)
}


crates_array_part_1 <- move_boxes(moves, crates_array,style = "sequential")
```

```r
print(crates_array_part_1)
lapply(crates_array_part_1, `[[`, 1)
```

```r
crates_array <- crates_array_in
crates_array_part_2 <- move_boxes(moves, crates_array, style = "all_in_one")
```

```r
print(crates_array_part_2)
lapply(crates_array_part_2, `[[`, 1)
```
